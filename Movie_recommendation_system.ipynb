# Text to vector 
## Textvectorization-- we are using BAGS OF WORDS technique

```python
# bag of word-->step1 : combines all movie tags (1 movie tag contains 1000 words then 2k movies contains 5lac words) 
# step2: out of total tags we can find 5000 common words which are using very frequently except stop words(are,in,of,etc)
# step3:then from 5000 words we take only 1 word and check in movie1 tags after that check in movie2 and so on to 5k movies
# then again take 2nd word and check in all movies tags and continues check all words in all movies(create 5000*5000 dimensions)
#shape should be 4806*5000
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=5000,stop_words="english")
```

```python
vectors=cv.fit_transform(movie["tags"]).toarray()
# 0 is showing because movie tags may not be contain 5000 words-->jb koi word nhi milega tag mein toh 0 hi aayega 
```

```python
vectors
```

### Now we find distance of every vector to each and every vector
#### if distance from one vector to another vector is more then similarity is low

```python
# we are using cosine distance instead of euclidean distance as it gives error on large dataset
# cosine similarity give value in the range (0 and 1) :1 means full similar movie and 0 means not similar 
# as decreases from 1 to 0 similarity decreases
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)  #-->gives range from 0 to 1 
# note:in this array diagonally value is always 1 ->because o movie ki similarity o movie se poori 1 hogi
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
# now sort in desc on the basis of distance of similarity array 
```

```python
def recommend(movie_title):
     if movie[movie["title"] == movie_title].empty:
         print(f"Movie '{movie_title}' not found in the dataset.")
         return
    
     movie_index=movie[movie["title"]== movie_title].index[0]
     distances=similarity[movie_index]
     movie_recommend_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
     for i in movie_recommend_list:
         print(movie.iloc[i[0]].title)
```
